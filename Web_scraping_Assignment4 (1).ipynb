```python
Repository_title = []
Repository_description = []
Contributors_count = []
Language_used = []
```

```python
try:
    tag1 = driver.find_elements(By.XPATH,'//h1[@class="h3 lh-condensed"]')
    for i in tag1:
        Repository_title.append(i.text)
except NoSuchElementException as e:
    Repository_title.append('-')
    
try:
    tag2 = driver.find_elements(By.XPATH,'//p[@class="col-9 color-fg-muted my-1 pr-4"]')
    for i in tag2:
        Repository_description.append(i.text)
except NoSuchElementException as e:
    Repository_description.append('-')
    
try:
    tag3 = driver.find_elements(By.XPATH,'//a[@class="Link--muted d-inline-block mr-3"][2]')
    for i in tag3:
        Contributors_count.append(i.text)
except NoSuchElementException as e:
    Contributors_count.append('-')
    
try:
    tag4 = driver.find_elements(By.XPATH,'//span[@class="d-inline-block ml-0 mr-3"]')
    for i in tag4:
        Language_used.append(i.text)
except NoSuchElementException as e:
    Language_used.append('-')
```

```python
print(len(Repository_title),len(Repository_description),len(Contributors_count),len(Language_used))
```

```python
df = pd.DataFrame({'Repository title':Repository_title[0:21],'Repository description':Repository_description[0:21],
                   'Contributors count':Contributors_count[0:21],'Language used':Language_used[0:21]})
df
```

```python
#7
driver = webdriver.Chrome(r'C:\New folder\webdriver.exe')
driver.get('http://www.imdb.com/list/ls095964455/')
```

```python
Name = []
Year_span = []
Genre = []
Run_time = []
Ratings = []
Votes = []
```

```python
tag1 = driver.find_elements(By.XPATH,'//h3[@class="lister-item-header"]')
for i in tag1:
    Name.append(i.text.split('.')[1].split('(')[0])
    
    
tag2 = driver.find_elements(By.XPATH,'//span[@class="lister-item-year text-muted unbold"]')
for i in tag2:
    Year_span.append(i.text)
    
tag3 = driver.find_elements(By.XPATH,'//span[@class="genre"]')
for i in tag3:
    Genre.append(i.text)
    
tag4 = driver.find_elements(By.XPATH,'//span[@class="runtime"]')
for i in tag4:
    Run_time.append(i.text)
    
tag5 = driver.find_elements(By.XPATH,'//div[@class="ipl-rating-star small"]')
for i in tag5:
    Ratings.append(i.text)
    
tag6 = driver.find_elements(By.XPATH,'//div[@class="lister-item-content"]')
for i in tag6:
    Votes.append(i.text.split('\n')[6].split(':')[1])
```

```python
print(len(Name),len(Year_span),len(Genre),len(Run_time),len(Ratings),len(Votes))
```

```python
df = pd.DataFrame({'Name':Name,'Year span':Year_span,'Genre':Genre,'Run time':Run_time,'Ratings':Ratings,'Votes':Votes})
df
```
