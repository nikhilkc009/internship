```python
#2
Brand_name = []
Name_of_the_product = []
price = []
Return_Exchange = []
Expected_delivery = []
Avaliability = []
Product_url = []
```

```python
start = 0
end = 2
for page in range(start,end):
    url_tags = driver.find_elements(By.XPATH,'//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
    for i in url_tags:
        Product_url.append(i.get_attribute('href'))
    next_button = driver.find_element(By.XPATH,'//a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]')
    next_button.click()
    time.sleep(3)
```

```python
for i in Product_url:
    driver.get(i)
    time.sleep(2)
    try:
        tag1 = driver.find_elements(By.XPATH,'//tr[@class="a-spacing-small po-brand"]')
        for e in tag1:
            Brand_name.append(e.text.split(' ')[1])
    except NoSuchElementException as e:
        Brand_name.append('-')
        
    try:
        tag2 = driver.find_elements(By.XPATH,'//span[@class="a-size-large product-title-word-break"]')
        for e in tag2:
            Name_of_the_product.append(e.text)
    except NoSuchElementException as e:
        Name_of_the_product.append('-')
    
     
    try:
        tag3 = driver.find_elements(By.XPATH,'//span[@class="a-price aok-align-center reinventPricePriceToPayMargin priceToPay"]')
        for e in tag3:
            price.append(e.text)
    except NoSuchElementException as e:
        price.append('-')
        
        
    try:
        tag4 = driver.find_elements(By.XPATH,'//div[@class="a-column a-span3 a-text-center icon-container icon-grid-width celwidget"]')
        for e in tag4:
            Return_Exchange.append(e.text)
    except NoSuchElementException as e:
        Return_Exchange.append('-')
        
        
    try:
        tag5 = driver.find_elements(By.XPATH,'//div[@class="a-spacing-base"]')
        for e in tag5:
            Expected_delivery.append(e.text)
    except NoSuchElementException as e:
        Expected_delivery.append('-')
        
        
    try:
        tag6 = driver.find_elements(By.XPATH,'//div[@class="a-section a-spacing-none a-spacing-top-micro }"]')
        for e in tag6:
            Avaliability.append(e.text)
    except NoSuchElementException as e:
        Avaliability.append('-')
```

```python
print(len(Brand_name),len(Name_of_the_product),len(price),len(Return_Exchange),len(Expected_delivery),len(Avaliability),len(Product_url))
```

```python
df = pd.DataFrame({'Brand Name':Brand_name[0:53],'Name of the Product':Name_of_the_product[0:53],'price':price[0:53],
                   'Return/Exchange':Return_Exchange[0:53],'Expected Delivery':Expected_delivery[0:53],
                   'Avaliability':Avaliability[0:53],'Product URL':Product_url[0:53]})
df
```

```python
comments = []
comment_upvote = []
time = []
```

```python
tag1 = driver.find_elements(By.XPATH,'//span[@class="style-scope ytd-comment-action-buttons-renderer"]')
for i in tag1[0:500]:
    comment_upvote.append(i.text)
    driver.execute_script('window.scrollBy(0,500)')
    
tag2 = driver.find_elements(By.XPATH,'//span[@class="style-scope yt-formatted-string"]')
for i in tag2[0:500]:
    comments.append(i.text)
    driver.execute_script('window.scrollBy(0,500)')
    
    
tag3 = driver.find_elements(By.XPATH,'//a[@class="yt-simple-endpoint style-scope yt-formatted-string"]')
for i in tag3[0:500]:
    time.append(i.text)
    driver.execute_script('window.scrollBy(0,500)')
```
