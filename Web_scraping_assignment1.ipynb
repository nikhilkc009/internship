```python
#2
page2 = requests.get('https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc')
```

```python
soup1 = BeautifulSoup(page2.content)
```

```python
name = []
for i in soup1.find_all('h3', class_="lister-item-header"):
    name.append(i.text.split('\n')[2])
```

```python
rating = [] 
for i in soup1.find_all('div',class_="inline-block ratings-imdb-rating"):
    rating.append(i.text.split('\n')[2])
```

```python
year = []
for i in soup1.find_all('span',class_="lister-item-year text-muted unbold"):
    year.append(i.text)
```

```python
df = pd.DataFrame({'name':name,'ratings':rating,'year of release':year})
df
```

```python
#5.a)
teams = requests.get('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
```

```python
top_points = []
for i in teams_soup.find_all('td',class_='rankings-block__banner--points'):
    top_points.append(i.text)
    for i in teams_soup.find_all('tr',class_='table-body'):
        top_points.append(i.text.split('\n')[8])
```

```python
df_teams = pd.DataFrame({'Teams':top_teams[:10],'Ratings':top_ratings[:10],'Matches':top_matches[:10],'Points':top_points[:10]})
df_teams
```

```python
#9
dineout = requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
```

```python
dineout_soup=BeautifulSoup(dineout.content)
```

```python
res_name=[]
for i in dineout_soup.find_all('a',class_='restnt-name ellipsis'):
    res_name.append(i.text)
```

```python
cus=[]
for i in dineout_soup.find_all('span','double-line-ellipsis'):
    cus.append(i.text.split('|')[1])
```

```python
loc=[]
for i in dineout_soup.find_all('div',class_='restnt-loc ellipsis'):
    loc.append(i.text)
```

```python
dine_ratings=[]
for i in dineout_soup.find_all('div',class_='restnt-rating rating-4'):
    dine_ratings.append(i.text)
```

```python
image=[]
for i in dineout_soup.find_all('img',class_='no-img'):
    image.append(i.get('data-src'))
```

```python
dineout_df = pd.DataFrame({'Restaurant Name':res_name,'Cuisine':cus,'Location':loc,'Ratings':dine_ratings,'Image':image})
dineout_df
```
